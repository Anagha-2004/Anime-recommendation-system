```python
import pandas as pd

# Load the dataset (adjust path if needed)
anime_df = pd.read_csv("anime.csv")

# Show the first few rows
anime_df.head()
```

```python
# Overview of columns and data types
anime_df.info()

# Count missing values in each column
anime_df.isnull().sum()
```

```python
# Summary stats for numeric columns
anime_df.describe()

# Unique values in categorical columns
print("\nUnique types:", anime_df['type'].unique())
print("\nUnique genres (first 5 rows):", anime_df['genre'].head())
```

```python
anime_df = anime_df.dropna(subset=['rating', 'genre'])
```

```python
# Check how many rows and columns remain after dropping missing values
anime_df.shape
```

```python
anime_df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
# Fill missing genres with empty strings
anime_df['genre'] = anime_df['genre'].fillna('')

# Initialize TF-IDF Vectorizer (ignore stop words, etc.)
tfidf = TfidfVectorizer(stop_words='english')

# Compute TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(anime_df['genre'])

# Shape of matrix: (num_anime x num_genres)
print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python
# Compute cosine similarity between all anime based on genre
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Reset index in case it’s not aligned
anime_df = anime_df.reset_index()

# Create a mapping from anime name to index
indices = pd.Series(anime_df.index, index=anime_df['name']).drop_duplicates()
```

```python
def recommend_anime(title, num_recommendations=5):
    if title not in indices:
        return f"'{title}' not found in dataset."

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    anime_indices = [i[0] for i in sim_scores]
    
    return anime_df[['name', 'genre', 'type']].iloc[anime_indices]
```

```python
# Combine genre and type into one string
anime_df['features'] = anime_df['genre'] + ' ' + anime_df['type'].fillna('')

# Recompute TF-IDF and similarity
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(anime_df['features'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

```

```python
result = recommend_anime("Naruto", num_recommendations=10)
result.to_csv("naruto_recommendations.csv", index=False)
```
